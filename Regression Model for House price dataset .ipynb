### Importing required libraries 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import warnings
warnings.filterwarnings(action= 'ignore')
```

```python
data = pd.read_csv('Transformed_Housing_Data2.csv')
data.head()
```

### Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
Y = data['Sale_Price']
X = scalar.fit_transform(data.drop(columns= ['Sale_Price']))
X = pd.DataFrame(data= X , columns=data.drop(columns= ['Sale_Price']).columns)
X.head()
```

### Checking and removing multicollinearity 

```python
X.corr()
```

```python
# pair of indpendent variables with corelation greater than 0.5
# defining a fuction to check for high correaltion  
k = X.corr()
z = [[str(i), str(j)] for i in k.columns for j in k.columns if(k.loc[i,j] > abs(0.5)) &(i!=j)]
z , len(z)
```

### Calculating VIF 

```python
# Treat multicollinearity with VIF
# importing Varience_infation_factor from the statsmodel
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = X

#  calculating VIF for every column
VIF = pd.Series([variance_inflation_factor(vif_data.values , i) for i in range (vif_data.shape[1])], index = vif_data.columns) 
VIF
```

```python
VIF[VIF == VIF.max()] .index[0]
```

```python
def MC_remover(data):
  vif = pd.Series([variance_inflation_factor(data.values, i) for i in range(data.shape[1])], index = data.columns)
  if vif.max() > 5:
    print(vif[vif == vif.max()].index[0],'has been removed')
    data = data.drop(columns = [vif[vif == vif.max()].index[0]])
    return data
  else:
    print('No Multicollinearity present anymore')
    return data
```

```python
for i in range(7):
    vif_data = MC_remover(vif_data)
    
vif_data.head()
```

```python
VIF = pd.Series([variance_inflation_factor(vif_data.values , i) for i in range (vif_data.shape[1])], index = vif_data.columns) 
VIF
```

# Remaining columns

# Train / Test set

```python
x = vif_data 
y = data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test  = train_test_split (X ,Y  ,test_size = 0.3 , random_state  = 101 )

x_train.shape  , x_test.shape  , y_train.shape  , y_test.shape
```

# Linear Regression 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True)
lr.fit(x_train ,y_train)
```

```python
lr.coef_
```

```python
prediction = lr.predict(x_test)
lr.score(x_test , y_test)
```

# 1. Residuals 

```python
residuals  = prediction - y_test

residuals_table  = pd.DataFrame({ 'residuals' : residuals ,'prediction' : prediction})

residuals_table = residuals_table.sort_values( by = 'prediction') 
```

```python
z = [i for i in range(int(residuals_table['prediction'].max()))]
k = [0 for i in range(int(residuals_table['prediction'].max()))]
```

```python
plt.figure(dpi = 130, figsize = (17,7))

plt.scatter( residuals_table['prediction'], residuals_table['residuals'], color = 'red', s = 2)
plt.plot(z, k, color = 'green', linewidth = 3, label = 'regression line')
plt.ylim(-800000, 800000)
plt.xlabel('fitted points (ordered by predictions)')
plt.ylabel('residuals')
plt.title('residual plot')
plt.legend()
plt.show()
```

# 2. Distribution of errors

```python
plt.figure(dpi = 100, figsize = (10,7))
plt.hist(residuals_table['residuals'], color = 'red', bins = 200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('distribution of residuals')
plt.show()
```

# Model Coefficients

```python
coefficients_table = pd.DataFrame({'column': x_train.columns,
                                  'coefficients': lr.coef_})
coefficient_table = coefficients_table.sort_values(by = 'coefficients')
```
